```python
import matplotlib.pyplot as plt
import numpy as np

# X-axis values: number of GPUs
x_values = [256, 512, 1024, 2048, 4096, 8192, 16384]

# Y-axis values: Model Flops Utilization (MFU)%
y_values_ours = [55 for _ in x_values]  # Constant 55%
y_values_megatron = np.linspace(50, 30, len(x_values))  # Decline from 50% to 30%

# Create the plot
plt.figure(figsize=(10, 6), dpi=300)

# Plot the "ours" line
plt.plot(x_values, y_values_ours, label='Ours', marker='o', linestyle='-', linewidth=2)

# Plot the "Megatron-LM" line
plt.plot(x_values, y_values_megatron, label='Megatron-LM', marker='x', linestyle='--', linewidth=2)

# Adding title and labels
plt.title(r'Fixed model size = $175B$, Scale #GPU', fontsize=16)
plt.xlabel('Number of GPUs', fontsize=14)
plt.ylabel('Model Flops Utilization (MFU)%', fontsize=14)

# Adding grid and setting axis limits
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim(0, 17000)
plt.ylim(20, 60)

# Adding legend with fontsize
plt.legend(fontsize=12)

# Save the plot as a pdf
plt.savefig("muf_vs_gpus.pdf")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# X-axis values: Different techniques
tech_labels = ['tech1', 'tech2', 'tech3', 'tech4', 'tech5', 'All']

# Y-axis values: Model Flops Utilization (MFU) Increase %
y_values = [4, 5, 3, 6, 7, 20]  # These are example percentages; adjust as necessary.

# Create the bar chart
plt.figure(figsize=(10, 6), dpi=300)

# Create the bars
bars = plt.bar(tech_labels, y_values, color=['blue', 'green', 'red', 'purple', 'orange', 'cyan'])

# Add data labels above the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.5, str(yval), ha='center', va='bottom', fontsize=10)

# Adding title and labels with larger fonts
plt.title(r'Model Flops Utilization (MFU) Increase % by Technique', fontsize=16)
plt.xlabel('Techniques', fontsize=14)
plt.ylabel('MFU Increase (%)', fontsize=14)

# Add grid lines on the y-axis for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust y-axis limits
plt.ylim(0, 25)  # Adjust as needed

# Save the bar chart as a high-resolution image
plt.savefig("mfu_inc_breakdown.pdf")

# Show the plot
plt.show()
```
